## Indexing/Slicing Rows and columns with loc[]

- dataframe.loc['row_name','column_name] <br>
- If several rows and columns: put in []


```python
summer.loc['PHELPS, Michael', 'Medal']
```

```python
summer.loc[['PHELPS, Michael', 'LEWIS, Carl'], ['Medal', 'Event' ]]
```

```python
summer.loc[:, ['Medal', 'Event' ]]
```

```python
summer.loc[:, ['Medal', 'Event' ]].equals(summer[['Medal','Event']])
```

```python
#Get all rows until CHASAPIS, Spiridon (include)>> should be unique 
summer.loc[:'CHASAPIS, Spiridon']
```
