## Manipulating Values in a DataFrame

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

###  (Option : loc)

```python
titanic.loc[1,'age'] = 40
```

```python
titanic.head()
```

```python
titanic.loc[1:3,'age'] = 50
```

```python
titanic.head()
```

### (Option: iloc)

```python
titanic.iloc[1,3] = 42
```

```python
titanic.head()
```

```python
titanic.iloc[1:4,3] = [43,44,45]
```

```python
titanic.head()
```

### (Option 3: boolean indexing)

```python
index_babies = titanic.loc[titanic.age < 1,'age']
index_babies
```

```python
index_babies = titanic.loc[titanic.age < 1].index
index_babies
```

```python
#Get the list of all indeices
index_babies = titanic.loc[titanic.age < 1,'age'].index
index_babies
```

```python
titanic.loc[titanic.age <1,'age'] = 1
```

```python
titanic.loc[index_babies]
```

## View vs. Copy

### Slicing a DataFrame / Creating a view on the original DataFrame

```python
import pandas as pd
```

```python
titanic  = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age
```

```python
age.head()
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
titanic.query('2 <age<=5')
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
df_baby
```

```python
df_baby._is_view
```

```python
df_baby._is_copy is None
```

```python
df_baby._is_copy()
```

## To manipulate a Slice of a DataFrame

- avoid chained Indexing <br>
- copy with .copy()

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age.copy()
```

```python
age.head()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
baby_ages = titanic.age[titanic.age < 1,['age','sex']].copy()
```
