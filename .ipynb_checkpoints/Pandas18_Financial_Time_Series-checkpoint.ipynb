## Measuring Stock Performance with MEAN Return and STD of Returns

```python
import numpy as np
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
aapl.pct_change().dropna()
```

```python
ret = aapl.pct_change().dropna()
ret.head()
```

```python
ret.info()
```

```python
ret.plot(kind='hist', figsize=(12,8), bins=100)
plt.show()
```

```python
daily_mean_Return = ret.mean()
daily_mean_Return
```

```python
var_daily_Return = ret.var()
var_daily_Return
```

```python
std_daily_Returns = np.sqrt(var_daily_Return)
std_daily_Returns
```

```python
ret.std()
```

```python
ann_var_Returns = ret.mean() * 252
ann_var_Returns
```

```python
ann_std_Returns = np.sqrt(ann_var_Returns)
ann_std_Returns
```

```python
ret.std() * np.sqrt(252)
```

## Financial Time Series - Return and Risk

```python
norm.plot(figsize=(15,8), fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
close.head()
```

```python
close.pct_change().dropna()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.describe().T.loc[:,['mean','std']]
```

```python
summary = ret.describe().T.loc[:,['mean','std']]
summary
```

```python
summary['mean'] = summary['mean']*252
summary['std'] = summary['std'] * np.sqrt(252)
```

```python
summary
```

```python
summary.plot(kind='scatter', x='std', y='mean', figsize=(15,12), s=50, fontsize=15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, 'std']+0.002, summary.loc[i, 'mean']+0.002),size=15)
plt.xlabel('ann. Risk (std)', fontsize=15)
plt.ylabel('ann. Return', fontsize=15)
plt.title('Risk/Return', fontsize=20)
plt.show()
```
