## The plot() method

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.tail()
```

```python
titanic.info()
```

```python
titanic.plot(subplots = True, figsize=(16,12), sharex = False, sharey = False)
plt.show()
```

```python
titanic.age.plot(figsize = (12,8))
plt.show()
```

## Customization of plots

```python
xticks = [x for x in range(0,901,50)]
xticks
```

```python
yticks = [y for y in range(0,81,5)]
yticks
```

```python
#Style sheets available (types of graphs)
plt.style.available
```

```python
plt.style.use('classic')
```

```python
titanic.age.plot(figsize = (12,8), fontsize = 13, c = 'r', linestyle = '-', xlim = (0,900), ylim = (0,80), xticks = xticks,
                yticks = yticks, rot = 45)
plt.title('Titanic - Ages', fontsize = 15)
plt.legend(loc=3, fontsize=15)
plt.xlabel('Passenger No', fontsize= 13)
plt.ylabel('Age', fontsize= 13)
#plt.grid
plt.show()
```

## Histograms

```python
plt.style.use('seaborn')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.age.value_counts()
```

```python
titanic.age.plot(kind = 'hist', figsize = (12,8), fontsize = 15, bins = 80, density = True)
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize = 15, ylabelsize = 15, cumulative = True)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative = True)
plt.show()
```
